```python
import pandas as pd 
import numpy as np

from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.speedtest.net/global-index'
response = requests.get(url)

# Ensure the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'lxml')

    # Create bsobj from the soup object
    bsobj = BeautifulSoup(str(soup), 'lxml')
    
    # Verify bsobj is created
    print(bsobj)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import pandas as pd

# Assuming countries, speeds, act_rank, and change_rank lists are already defined
countries = soup.find_all('td', class_='country')
speeds = soup.find_all('td', class_='speed')
act_rank = soup.find_all('td', class_='rank actual-rank')
change_rank = soup.find_all('td', class_='rank rank-change')

# Check if the lengths of the lists match
if len(countries) == len(speeds) == len(act_rank) == len(change_rank):
    data = []
    for country, speed, act_r, change_r in zip(countries, speeds, act_rank, change_rank):
        # Collect country, speed, act_rank, and change_rank into a list of dictionaries
        data.append({
            'Country': country.get_text().strip(),
            'Speed': speed.get_text().strip(),
            'Act_rank': act_r.get_text().strip(),
            'Chag_rank': change_r.get_text().strip()
        })
    
    # Convert the list of dictionaries into a DataFrame
    df = pd.DataFrame(data)
    print(df)
else:
    print("The number of country, speed, actual rank, and change rank elements do not match.")
```
